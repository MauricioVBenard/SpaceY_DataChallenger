```python
import pandas as pd
import requests
import io
import seaborn as sns
import matplotlib.pyplot as plt

URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"

try:
    response = requests.get(URL)
    response.raise_for_status()
    dataset_part_2_csv = io.StringIO(response.text)
    df = pd.read_csv(dataset_part_2_csv)

    # Function to Extract years from the date
    year = []
    def Extract_year():
        for i in df["Date"]:
            year.append(i.split("-")[0])
        return year

    Extract_year()
    df['Date'] = year

    # Calculate success rate per year
    df['Class'] = df['Class'].astype(int) #make sure class column is an integer.
    success_rate = df.groupby('Date')['Class'].mean()

    # Plot a line chart with x axis to be the extracted year and y axis to be the success rate
    plt.figure(figsize=(10, 6))
    plt.plot(success_rate.index, success_rate.values, marker='o')
    plt.xlabel("Year", fontsize=15)
    plt.ylabel("Success Rate", fontsize=15)
    plt.title("Success Rate per Year", fontsize=16)
    plt.grid(True)
    plt.xticks(rotation=45) #rotate the x axis ticks so they are easier to read.
    plt.show()

except requests.exceptions.RequestException as e:
    print(f"Error fetching data: {e}")
except pd.errors.ParserError as e:
    print(f"Error parsing CSV: {e}")
```

```python
import pandas as pd
import requests
import io

URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"

try:
    response = requests.get(URL)
    response.raise_for_status()
    dataset_part_2_csv = io.StringIO(response.text)
    df = pd.read_csv(dataset_part_2_csv)

    # Identify categorical columns
    categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
    print(f"Categorical columns before get_dummies: {categorical_cols}")

    # Create dummy variables for categorical columns
    df_with_dummies = pd.get_dummies(df, columns=categorical_cols, drop_first=True) #drop_first=True

    # Identify numeric columns
    numeric_cols = df_with_dummies.select_dtypes(include=['number']).columns.tolist()
    print(f"\nNumeric columns before casting: {numeric_cols}")

    # Cast all numeric columns to float64
    df_with_dummies[numeric_cols] = df_with_dummies[numeric_cols].astype('float64')

    # Verify the data types of the columns
    print("\nData types of the DataFrame:")
    print(df_with_dummies.dtypes)

    # Display the first few rows of the DataFrame
    print("\nDataFrame with numeric columns cast to float64:")
    print(df_with_dummies.head())

    # Export to CSV
    df_with_dummies.to_csv('spacex_dataset.csv', index=False)
    print("\nDataFrame successfully exported to spacex_dataset.csv")

except requests.exceptions.RequestException as e:
    print(f"Error fetching data: {e}")
except pd.errors.ParserError as e:
    print(f"Error parsing CSV: {e}")

# Note for the next lab:  The data in 'spacex_dataset.csv' may be in a pre-selected date range for the next lab.
```
