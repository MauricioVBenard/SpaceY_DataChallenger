```python
def date_time(table_cells):
    """
    This function returns the data and time from the HTML  table cell
    Input: the  element of a table data cell extracts extra row
    """
    return [data_time.strip() for data_time in list(table_cells.strings)][0:2]

def booster_version(table_cells):
    """
    This function returns the booster version from the HTML  table cell 
    Input: the  element of a table data cell extracts extra row
    """
    out=''.join([booster_version for i,booster_version in enumerate( table_cells.strings) if i%2==0][0:-1])
    return out

def landing_status(table_cells):
    """
    This function returns the landing status from the HTML table cell 
    Input: the  element of a table data cell extracts extra row
    """
    out=[i for i in table_cells.strings][0]
    return out


def get_mass(table_cells):
    mass=unicodedata.normalize("NFKD", table_cells.text).strip()
    if mass:
        mass.find("kg")
        new_mass=mass[0:mass.find("kg")+2]
    else:
        new_mass=0
    return new_mass


def extract_column_from_header(row):
    """
    This function returns the landing status from the HTML table cell 
    Input: the  element of a table data cell extracts extra row
    """
    if (row.br):
        row.br.extract()
    if row.a:
        row.a.extract()
    if row.sup:
        row.sup.extract()
        
    colunm_name = ' '.join(row.contents)
    
    # Filter the digit and empty names
    if not(colunm_name.strip().isdigit()):
        colunm_name = colunm_name.strip()
        return colunm_name    
```

```python
import requests
from bs4 import BeautifulSoup
import re

static_url = "https://en.wikipedia.org/w/index.php?title=List_of_Falcon_9_and_Falcon_Heavy_launches&oldid=1027775405"
response = requests.get(static_url)
html_content = response.text  # Save the raw HTML

# Print the beginning of the raw HTML
print("Beginning of raw HTML:")
print(html_content[:1000])  # Print the first 1000 characters

# Check if "Flight No." is in the raw HTML
if "Flight No." in html_content:
    print("The string 'Flight No.' is present in the raw HTML.")
else:
    print("The string 'Flight No.' is NOT present in the raw HTML.")

soup = BeautifulSoup(html_content, 'html.parser')
html_tables = soup.find_all('table', "wikitable plainrowheaders collapsible")

column_names = []

def extract_column_from_header(th):
    """Extracts the column name from a table header (th) element."""
    name = ""
    if th.a:
        name = th.a.text.strip()
    elif th.span:
        name = th.span.text.strip()
    else:
        name = th.text.strip()
    return name

# Define the unique string to search for
unique_string = "Flight No."

# Find the element containing the unique string
string_element = soup.find(string=re.compile(unique_string))

if string_element:
    # Navigate to the parent table element
    first_launch_table = string_element.find_parent("table")
    print("Found the table using unique string")
else:
    first_launch_table = None
    print("Error: Could not find the correct table on the page.")
    print("Dumping first 5 tables for inspection:")
    html_tables = soup.find_all('table', "wikitable plainrowheaders collapsible")
    for i, table in enumerate(html_tables[:5]):
        print(f"Table {i}:")
        print(table.prettify()[:2000])  # Print the first 2000 characters of each table
    # Consider adding code to handle this case, e.g., by exiting.
    column_names = [] # set to empty list to avoid errors later
    

if first_launch_table: #check if a table was found
    for th in first_launch_table.find_all('th'):
        name = extract_column_from_header(th)
        if name is not None and len(name) > 0:
            column_names.append(name)
else:
    print("Error: Could not find a suitable table on the page.")
    column_names = [] # set to empty list to avoid errors later

launch_dict = dict.fromkeys(column_names)
if 'Date and time ( )' in launch_dict:
    del launch_dict['Date and time ( )']
launch_dict['Flight No.'] = []
launch_dict['Launch site'] = []
launch_dict['Payload'] = []
launch_dict['Payload mass'] = []
launch_dict['Orbit'] = []
launch_dict['Customer'] = []
launch_dict['Launch outcome'] = []
launch_dict['Version Booster'] = []
launch_dict['Booster landing'] = []
launch_dict['Date'] = []
launch_dict['Time'] = []


def date_time(table_cells):
    """Extracts date and time from the table cells."""
    return [data_time.strip() for data_time in table_cells.text.split('\n')]

def booster_version(table_cells):
    """Extracts the booster version from the table cells"""
    bv = table_cells[1].a.string
    if not bv:
        bv = table_cells[1].a.text
    return bv

def get_mass(table_cells):
    """Extracts the payload mass from the table cells."""
    mass = table_cells.text.strip()
    if mass:
        mass = ''.join(mass.split(sep=' ')[0].split(','))
    return mass

def landing_status(table_cells):
    """Extracts the landing status from the table cells."""
    outcome = list(table_cells.strings)[0].strip()
    return outcome



extracted_row = 0
# Extract each table
if first_launch_table: # only extract if a table was found.
    for table_number, table in enumerate(soup.find_all('table', "wikitable plainrowheaders collapsible")):
        if table == first_launch_table: #make sure we are using the table we found
            # get table row
            for rows in table.find_all("tr"):
                # check to see if first table heading is as number corresponding to launch a number
                if rows.th:
                    if rows.th.string:
                        flight_number = rows.th.string.strip()
                        flag = flight_number.isdigit()
                else:
                    flag = False
                # get table element
                row = rows.find_all('td')
                # if it is number save cells in a dictionary
                if flag:
                    extracted_row += 1
                    try:
                        datatimelist = date_time(row[0])
                        date = datatimelist[0].strip(',')
                        time = datatimelist[1]
                        bv = booster_version(row[1])
                        launch_site = row[2].a.string
                        payload = row[3].a.string
                        payload_mass = get_mass(row[4])
                        orbit = row[5].a.string
                        customer = row[6].a.string
                        launch_outcome = list(row[7].strings)[0].strip()
                        booster_landing = landing_status(row[8])


                        launch_dict['Flight No.'].append(flight_number)
                        launch_dict['Date'].append(date)
                        launch_dict['Time'].append(time)
                        launch_dict['Version Booster'].append(bv)
                        launch_dict['Launch site'].append(launch_site)
                        launch_dict['Payload'].append(payload)
                        launch_dict['Payload mass'].append(payload_mass)
                        launch_dict['Orbit'].append(orbit)
                        launch_dict['Customer'].append(customer)
                        launch_dict['Launch outcome'].append(launch_outcome)
                        launch_dict['Booster landing'].append(booster_landing)
                    except IndexError as e:
                        print(f"IndexError at row {extracted_row}: {e}")
                        print(f"Row data: {row}")
                        launch_dict['Flight No.'].append(None)
                        launch_dict['Date'].append(None)
                        launch_dict['Time'].append(None)
                        launch_dict['Version Booster'].append(None)
                        launch_dict['Launch site'].append(None)
                        launch_dict['Payload'].append(None)
                        launch_dict['Payload mass'].append(None)
                        launch_dict['Orbit'].append(None)
                        launch_dict['Customer'].append(None)
                        launch_dict['Launch outcome'].append(None)
                        launch_dict['Booster landing'].append(None)

# Print the filled launch_dict to verify
print(launch_dict)
```
